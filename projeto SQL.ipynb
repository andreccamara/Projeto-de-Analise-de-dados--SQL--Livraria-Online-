## avaliaçõese classificação media <a id='media'></a>

Encontre o número de avaliações e a classificação média para cada livro.
para tanto usaremos duas tabelas books e rating


```python
consulta(books)
```

```python
consulta(ratings)
```

```python
# 1 o número de avaliações e 2 a classificação média para cada livro.
classificacao = """
SELECT books.book_id,count(ratings.rating) as numero_de_classificacao, ROUND(AVG(ratings.rating), 1) as classificacao_media, title as livro 
FROM books 
JOIN ratings 
ON books.book_id=ratings.book_id 
GROUP BY books.book_id
ORDER BY numero_de_classificacao DESC, classificacao_media DESC """
consulta(classificacao)
```

```python
avaliacao= f''' 
SELECT count(r.text), AVG(classificacao.classificacao_media)
from reviews r 
join({classificacao}) as classificacao 
on classificacao.book_id=r.book_id'''
consulta(avaliacao)
```

ok encontramos os numeros de avaliações 2793 e a media das medias de classificações por livro, sendo 3,9 . Além disso, é 6,46 a quantidade de avaliações por livro. as notas vão de 0 a 5, sendo que os unicos que contem 5 estrelas têm menos de 5 avaliações, já os 3 que foram mais classificados têm 160, depois 88 e 86 com classificações proximas a 4

o codigo a seguir é muito complicado então vou explicar:
    na primeira variavel eu armazenei uma junção de tabelas livro e avaliação para calcular a quantidade de avaliações -count(r.rating)- e separar apenas os com mais de 50.
    na segunda variavel eu juntei as tres tabelas livros autores e avaliações juntamente com a consulta da primeira variavel, assim, garanti acesso a quantidade de avaliações, depois filtrei por avalliações >=50 e agrupei por autor calculando a media de avaliações e ordenei em ordem decrescente

```python
livro_avaliacoes = 'SELECT title, COUNT(r.rating) AS numero_avaliacoes FROM books AS b JOIN ratings AS r ON r.book_id = b.book_id GROUP BY b.title'
```

```python
x=f'select* FROM authors a JOIN books b ON a.author_id=b.author_id JOIN ratings AS r ON r.book_id = b.book_id JOIN  ({livro_avaliacoes}) as livro50 ON b.title = livro50.title where numero_avaliacoes>=50'
consulta(x)
```

```python
media_avaliacoes = f'SELECT a.author, AVG(r.rating) as avaliacao_media FROM authors a JOIN books b ON a.author_id=b.author_id JOIN ratings AS r ON r.book_id = b.book_id JOIN  ({livro_avaliacoes}) as livro50 ON b.title = livro50.title where numero_avaliacoes>=50 GROUP BY a.author ORDER BY avaliacao_media DESC'
#media_avaliacao_autor=f' select author, avg(rating) from {contando_avaliacoes} as media_avaliacao group by title '
consulta(media_avaliacoes)
```

contagemusuario = f'SELECT re.username, count(re.text) FROM reviews re JOIN ({usuarios50mais}) AS u50 ON u50.username = re.username WHERE u50.count >= 50 GROUP BY re.username'

```python
usuarios50mais = 'SELECT username, count(r.rating) FROM ratings r GROUP BY username'

contagemusuario = f'''
    SELECT re.username, count(re.text) 
    FROM reviews re 
    JOIN ({usuarios50mais}) AS u50 ON u50.username = re.username 
    WHERE u50.count >= 50 
    GROUP BY re.username
'''
consulta(contagemusuario)
```

```python
mediausuario = f'''
    SELECT AVG(count) 
    FROM (
        {contagemusuario}
    ) AS subconsulta
'''
consulta(mediausuario)
```
